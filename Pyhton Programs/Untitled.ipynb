```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(100,10))
data = pd.read_csv('Headbrain.csv')
print(data.shape)
data.head()
```

```python
x=data['Head size'].values
y=data['Brain Weights'].values

mean_x=np.mean(x)
mean_y=np.mean(y)
length = len(x)

numer = 0
denom = 0
for i in range(length):
    numer += (x[i] - mean_x)*(y[i] - mean_y)
    denom += (x[i] - mean_x)**2
m = numer/denom
c = mean_y-(m*mean_x)
print(m,c)
```

```python
y_pred = []
for i in x:
    y_pred.append(m*x+c)
```
